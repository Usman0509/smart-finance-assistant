```python
import pandas as pd
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant.

    Args:
        file_path: Path to CSV file

    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    try:
        # Load the CSV file
        df = pd.read_csv(file_path)

        # Check that the required columns exist
        required_columns = ["Date", "Amount", "Category", "Description"]

        for column in required_columns:
            if column not in df.columns:
                raise ValueError(f"Missing required column: {column}")

        # Clean the Amount column by removing dollar signs
        df["Amount"] = df["Amount"].astype(str).str.replace("$", "", regex=False)

        # Convert Amount column to numeric values
        df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

        # Convert Date column to date format
        df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

        # Remove rows with missing or invalid important values
        df = df.dropna(subset=["Date", "Amount", "Category", "Description"])

        return df

    except Exception as e:
        print(f"Error loading transaction data: {e}")
        return None


# Test your function using sample data
sample_transactions = {
    "Date": ["2024-08-01", "2024-08-02", "2024-08-03", "2024-08-04", "2024-08-05"],
    "Amount": ["$45.50", "$12.00", "$89.95", "$3.50", "-$25.00"],
    "Category": ["Groceries", "Transport", "Entertainment", "Coffee", "Refund"],
    "Description": ["Woolworths", "Opal Card", "Concert Tickets", "Campus Cafe", "Returned Item"]
}

df_sample = pd.DataFrame(sample_transactions)

df_sample.to_csv("sample_transactions.csv", index=False)

clean_data = load_and_clean_transaction_data("sample_transactions.csv")
print(clean_data)
```

```python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights.

    Args:
        df: Cleaned transaction DataFrame

    Returns:
        dict: Analysis results and insights
    """

    # Check if data exists
    if df is None or df.empty:
        return None

    # Only use positive amounts for spending analysis
    # This avoids refunds or negative values affecting spending totals
    spending_df = df[df["Amount"] > 0].copy()

    # Calculate total spending
    total_spending = spending_df["Amount"].sum()

    # Calculate average transaction amount
    average_transaction = spending_df["Amount"].mean()

    # Calculate total spending by category
    category_totals = spending_df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

    # Calculate percentage of total spending for each category
    category_percentages = (category_totals / total_spending * 100).round(2)

    # Identify the top spending category
    top_category = category_totals.index[0] if not category_totals.empty else "None"

    # Store all analysis results in a dictionary
    analysis = {
        "total_spending": total_spending,
        "average_transaction": average_transaction,
        "category_totals": category_totals,
        "category_percentages": category_percentages,
        "top_category": top_category
    }

    return analysis


# Test your function
analysis = analyze_spending_patterns(clean_data)

print("Analysis Results")
print("=" * 40)

if analysis is not None:
    print(f"Total spending: ${analysis['total_spending']:.2f}")
    print(f"Average transaction: ${analysis['average_transaction']:.2f}")
    print(f"Top spending category: {analysis['top_category']}")
    print("\nSpending by category:")
    print(analysis["category_totals"])
    print("\nCategory percentages:")
    print(analysis["category_percentages"])
else:
    print("No analysis available.")
```

```python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations



def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis.

    Args:
        analysis_data: Dictionary with spending analysis results

    Returns:
        str: Formatted recommendations report
    """

    # Check if analysis data exists
    if analysis_data is None:
        return "No analysis data available. Please upload and analyse transaction data first."

    # Get values from the analysis dictionary
    total_spending = analysis_data["total_spending"]
    average_transaction = analysis_data["average_transaction"]
    top_category = analysis_data["top_category"]
    category_totals = analysis_data["category_totals"]
    category_percentages = analysis_data["category_percentages"]

    # Create a report as a list of lines
    report = []

    report.append("Financial Recommendations Report")
    report.append("=" * 45)
    report.append(f"Total spending: ${total_spending:.2f}")
    report.append(f"Average transaction amount: ${average_transaction:.2f}")
    report.append(f"Highest spending category: {top_category}")
    report.append("")

    report.append("Spending breakdown:")
    for category, amount in category_totals.items():
        percentage = category_percentages[category]
        report.append(f"- {category}: ${amount:.2f} ({percentage:.2f}% of spending)")

    report.append("")
    report.append("Recommendation:")

    if top_category != "None":
        report.append(f"Your highest spending area is {top_category}.")
        report.append(f"Consider reviewing your {top_category} expenses first because this category has the biggest impact on your budget.")
    else:
        report.append("No major spending category was found.")

    if average_transaction > 50:
        report.append("Your average transaction amount is relatively high, so reducing a few larger purchases could help you save money.")
    else:
        report.append("Your average transaction amount is fairly low, so tracking repeated small purchases may help identify savings opportunities.")

    report.append("")
    report.append("This advice is educational and should not replace professional financial advice.")

    return "\n".join(report)


# Test your function
recommendations = generate_financial_recommendations(analysis)
print(recommendations)
```

## RAG System for Financial Documents

```python
# 🤖 AI Collaboration: Document Retrieval Setup
# Simple RAG-style system using transaction summaries as retrieved context

from hands_on_ai.chat import get_response

def setup_financial_rag(df, user_question):
    """
    Simple RAG-style function for the Smart Finance Assistant.

    This function retrieves useful context from the user's transaction data,
    then sends that context with the user's question to the AI chatbot.

    Args:
        df: Cleaned transaction DataFrame
        user_question: User's finance question

    Returns:
        str: AI response using retrieved transaction context
    """

    if df is None or df.empty:
        return "No transaction data available for retrieval."

    # Retrieve context from transaction data
    analysis = analyze_spending_patterns(df)
    recommendations = generate_financial_recommendations(analysis)

    prompt = f"""
    You are Budget Buddy, a Smart Finance Assistant.

    Use the retrieved transaction context below to answer the user's question.

    Retrieved transaction context:
    {recommendations}

    User question:
    {user_question}

    Give a clear, practical, educational answer.
    Do not give professional financial advice.
    """

    try:
        response = get_response(prompt)
        return response
    except Exception as e:
        return f"RAG connection error: {e}"


# Test your RAG system
rag_answer = setup_financial_rag(
    clean_data,
    "What's a good budgeting strategy for someone who overspends on entertainment?"
)

print(rag_answer)
```

## Custom Financial Tools

```python
# 🤖 AI Collaboration: Custom Tool Development
# Custom financial tool: Savings Goal Calculator

def savings_goal_calculator(current_savings, target_amount, monthly_contribution):
    """
    Calculate how long it will take to reach a savings goal.

    Args:
        current_savings: Amount the user has already saved
        target_amount: The savings goal amount
        monthly_contribution: How much the user saves each month

    Returns:
        str: User-friendly savings goal result
    """

    # Input validation
    if target_amount <= 0:
        return "Target amount must be greater than zero."

    if current_savings < 0:
        return "Current savings cannot be negative."

    if monthly_contribution <= 0:
        return "Monthly contribution must be greater than zero."

    # Calculate remaining amount
    remaining_amount = target_amount - current_savings

    # Check if goal is already reached
    if remaining_amount <= 0:
        return "You have already reached your savings goal."

    # Calculate months needed
    months_needed = remaining_amount / monthly_contribution

    return f"You need approximately {months_needed:.1f} months to reach your savings goal."


# Test your custom financial tool
savings_result = savings_goal_calculator(
    current_savings=500,
    target_amount=2000,
    monthly_contribution=250
)

print(savings_result)
```

## Gradio UI Integration

```python
# 🤖 AI Collaboration: Professional UI Design
# Gradio interface for the Smart Finance Assistant

import gradio as gr

def run_finance_assistant(file, question, current_savings, target_amount, monthly_contribution):
    """
    Main function used by the Gradio interface.
    It loads transaction data, analyses spending, gives recommendations,
    answers a finance question, and runs the savings calculator.
    """

    if file is None:
        return "Please upload a CSV file first."

    # Load and clean uploaded transaction data
    df = load_and_clean_transaction_data(file.name)

    if df is None:
        return "Could not load the transaction data. Please check your CSV format."

    # Analyse spending
    analysis = analyze_spending_patterns(df)
    recommendations = generate_financial_recommendations(analysis)

    # Run savings calculator tool
    savings_result = savings_goal_calculator(
        current_savings,
        target_amount,
        monthly_contribution
    )

    # Use RAG-style transaction context for chatbot response
    if question:
        chat_answer = setup_financial_rag(df, question)
    else:
        chat_answer = "No finance question was entered."

    # Final formatted output
    final_output = f"""
SMART FINANCE ASSISTANT REPORT
==============================

SPENDING ANALYSIS
-----------------
{recommendations}

SAVINGS GOAL TOOL
-----------------
{savings_result}

CHATBOT RESPONSE
----------------
{chat_answer}
"""

    return final_output


def create_finance_assistant_ui():
    """
    Create a Gradio interface for the Smart Finance Assistant.
    """

    demo = gr.Interface(
        fn=run_finance_assistant,
        inputs=[
            gr.File(label="Upload transaction CSV"),
            gr.Textbox(
                label="Ask a finance question",
                placeholder="Example: How can I reduce my spending?"
            ),
            gr.Number(label="Current savings", value=500),
            gr.Number(label="Target savings amount", value=2000),
            gr.Number(label="Monthly contribution", value=250)
        ],
        outputs=gr.Textbox(label="Smart Finance Assistant Output", lines=20),
        title="Budget Buddy: Smart Finance Assistant",
        description="Upload transaction data, analyse spending, ask finance questions, and calculate savings goals."
    )

    return demo


# Launch your complete application
demo = create_finance_assistant_ui()
demo.launch()
```

The main technical skills used in this project were Python functions, pandas data processing, basic AI chatbot integration, a RAG-style transaction context system, custom financial tools, Gradio interface design, and testing.

The project is useful because many students and everyday users struggle to understand their spending habits. By turning transaction data into clear summaries and recommendations, the assistant helps users become more aware of their money habits and find possible areas for saving.

AI was used as a development partner throughout the project. It helped with understanding the assignment, creating functions, improving testing, and designing the Gradio interface. I reviewed and adapted the AI-generated suggestions so they matched my project requirements.

Overall, this notebook demonstrates how AI-assisted programming can be used to build a practical finance application that solves a real personal budgeting problem.
